```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'http://california.bizhwy.com/businesses.php?c=349'

# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# Examine results, determine element that contains sought info, store info in a list
results = soup.find_all('div', class_='biz')

bizNameList = []
addressList = []
cityStateZip = []

# grab name, address, & city, state, zip
for result in results:
    #Biz Name
    bizNameResult = result.find('b').text
    bizNameList.append(bizNameResult)
    
    # Address
    addressResult = result.find('br').nextSibling
    addressList.append(addressResult)
    
    #city state
    cityStateZipResult = result.findAll('br')[1].nextSibling
    cityStateZip.append(cityStateZipResult)
```

```python
# 2nd Div - phone & website
results = soup.find_all('div', class_='biz')

phoneList = []
websiteList = []

for result in results:
    #Phone
    secondDiv = result.findAll('div')[1]
    phoneNumber = secondDiv.find('br').previous_sibling
    phoneList.append(phoneNumber)
```

```python
#create dataframe of results
bizInfoDF = pd.DataFrame({"Business Name":bizNameList,"Address":addressList, "City State Zip":cityStateZip, "Phone":phoneList})

bizInfoDF["Phone"] = bizInfoDF["Phone"].str.replace("Phone:","")

bizInfoDF.to_csv('business_info.csv', encoding='utf-8')
bizInfoDF
```
